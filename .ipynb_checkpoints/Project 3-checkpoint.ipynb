```python
# Import dependencies
import pandas as pd
import numpy as np
#import matplot.lib.pyplot as plt
from scipy.stats import linregress
from pathlib import Path
```

```python
#Import csv file
csv_path = Path("data_science_job.csv")

# Read the CSV into a Pandas DataFrame
data_analysis_df = pd.read_csv(csv_path)

```

```python
# Create a Dataframe
data_analysis_df.head()
```

```python
# Cleaning the Data - remove the columns "salary_currency" and "salary column"
data_analysis_df.drop(['salary_currency', 'salary'], axis=1)
```

```python
# Rename Columns
data_renamed_df = data_analysis_df.rename(columns={'salary_in_usd':'salary'})
```

```python
# Rename data within columns for clarity
replacements = {'CT': 'Contract', 'FT': 'Full Time', 'PT': 'Part Time', 'FL': 'Freelance'}
data_renamed_df['employment_type'] = data_renamed_df['employment_type'].replace(replacements)
data_renamed_df.head()
```
